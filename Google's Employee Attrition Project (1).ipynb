#  Data Visualization 

```python
#visulaize data to check distributions of data via histogram
import matplotlib.pyplot as plt

# Select variables for univariate analysis
variables = ['Total Hours', 'Utilization%', 'Work Hours', 'Leave Hours', 'Training Hours', 'BD Hours', 'NC Hours']

# Plot histograms for each variable
plt.figure(figsize=(15, 8))
for i, variable in enumerate(variables, 1):
    plt.subplot(3, 3, i)
    plt.hist(result[variable], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {variable}')
    plt.xlabel(variable)
    plt.ylabel('Frequency')
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define the pairs of variables for scatter plots
variables = [
    ('Total Hours', 'Utilization%'),
    ('Work Hours', 'Leave Hours'),
    ('Training Hours', 'BD Hours'),
    ('Utilization%', 'NC Hours')
]

# Plot scatter plots
for x_var, y_var in variables:
    if x_var in result.columns and y_var in result.columns:
        plt.figure(figsize=(6, 4))  # Set size of the figure
        sns.scatterplot(x=x_var, y=y_var, data=result, color='skyblue')  # Plot with x and y variables
        plt.title(f'Scatter Plot of {x_var} vs {y_var}')  # Title of the plot
        plt.xlabel(x_var)  # X-axis label
        plt.ylabel(y_var)  # Y-axis label
        plt.grid(True)
        plt.show()
    else:
        print(f"Columns {x_var} or {y_var} not found in the dataset.")
```

```python
#check corrleation of data 
# Calculate correlation matrix
correlation_matrix = result.corr()

# Display correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
#visulaize correlation to identify high correlation variable and drop them
import seaborn as sns

# Plot correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

### Distribution Charts:

#### Distribution of Total Hours:

There is a concentration of employees around specific total hours, with a spike at the maximum of 4000 hours. This could represent employees who have reached a certain threshold in total working hours, perhaps due to full-time work or tenure.

#### Distribution of Utilization%:

The majority of employees are clustered around the lower end of the utilization spectrum (0-100%), with a few outliers above 100%. This suggests that most employees are operating at or below the expected utilization level, with only a few significantly exceeding it.

#### Distribution of Work Hours:

Work hours are dispersed, but there's a noticeable spread, with some employees accumulating significantly higher work hours. This might reflect differences in job roles or overtime.

#### Distribution of Leave Hours:

The majority of employees take fewer leave hours, with the distribution tailing off as the number of leave hours increases. This is typical in most companies, where fewer employees take extended leaves.
#### Distribution of Training Hours:

Training hours are skewed towards the lower end, with a majority of employees receiving little to moderate training. The tail on the right indicates that only a small number of employees undergo significant training.

#### Distribution of BD Hours and NC Hours:

Both BD (Business Development) and NC (Non-Contractual) hours are concentrated at the lower end of the spectrum. This suggests that only a few employees engage in business development activities, while the bulk of employees are not heavily involved in non-contractual work.

### Scatter Plots:

#### Scatter Plot of Total Hours vs Utilization%:

There’s a concentration of data points at the lower end of both total hours and utilization%. Employees with higher total hours tend to cluster within a certain utilization range, possibly indicating that those who work more are not necessarily the most utilized.

#### Scatter Plot of Work Hours vs Leave Hours:

As expected, employees with fewer work hours generally take fewer leave hours. There are some outliers who, despite having worked many hours, have taken significant leave.

#### Scatter Plot of Training Hours vs BD Hours:

There is no strong correlation between training hours and BD hours. Most employees have low BD hours regardless of their training, indicating that business development may not be directly linked to training efforts.

#### Scatter Plot of Utilization% vs NC Hours:

There is a clustering at low NC hours with low utilization percentages, suggesting that those involved in NC hours are typically underutilized. The scatter spreads further as utilization% increases, but NC hours remain relatively low.

## Key Insights:
#### Utilization Focus: The majority of employees are underutilized, which could be an area of focus for improvement, especially if the target utilization is higher than the distribution suggests.
####  Training and Development: Most employees receive limited training, which may impact long-term productivity and 
#### retention. Increasing training hours could help develop skills and increase utilization.
#### Leave Management: While leave hours are concentrated on the lower side, there are outliers taking substantial leave. It might be worth investigating whether certain groups of employees are at higher risk for burnout or are not being provided adequate support.
#### These charts collectively highlight the potential inefficiencies in utilization, training, and work allocation, which could inform future actions to optimize employee performance and retention.
